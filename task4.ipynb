Plot a bar graph to present the number of recipes written by each author. 

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
# Loading the json files 
data = pd.read_json('recipes.json', lines = True)
data.head(5)

# Ensure all author names are strings and clean the data
data['Author'] = data['Author'].astype(str)
```

```python
# Count the number of recipes by each author
author_counts = {}

# Iterate through the DataFrame rows
for index, row in data.iterrows():
    author = row["Author"]
    # checks if the author's name is already a key in the author_counts dictionary
    if author in author_counts:
        author_counts[author] += 1
    # if the authors name is not a key in the dictionary, the count in the dictionary 
    else:
        author_counts[author] = 1
        

# Sort the dictionary by values (number of recipes) in descending order
sorted_author_counts = dict(sorted(author_counts.items(), key=lambda item: item[1], reverse=True))

# Plot the bar graph
authors = list(sorted_author_counts.keys())
counts = list(sorted_author_counts.values())

plt.figure(figsize=(10, 6))
plt.bar(authors, counts, color='red')
plt.xlabel('Authors')
plt.ylabel('Number of Recipes')
plt.title('Number of Recipes Written by Each Author')
plt.xticks(rotation=90, fontsize=8)  # Make the font size smaller to avoid touching labels
plt.gca().margins(x=0.05)  # Add margins to the x-axis to further avoid label collision
plt.tight_layout()  # Adjust layout to make room for x-axis labels
plt.show()
```
