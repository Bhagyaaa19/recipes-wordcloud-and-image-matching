```python
# importing the packages
import json
import time


# Initialize an empty list to hold all recipes
recipes = []

# Read the JSON file line by line
with open('recipes.json', 'r') as file:
    for line in file:
        if line.strip():  # Ensure the line is not empty
            recipes.append(json.loads(line))

```

```python
# Extract recipe names and descriptions, ensuring descriptions are not None
extracted_data = [(recipe['Name'], recipe['Description']) for recipe in recipes if recipe['Description'] is not None]

print(extracted_data)
```

```python
keywords = ["kid", "kids", "child", "children","play","fun"]
recommended_foods = []

# Using nested loops
for name, description in recipe_data:
    for word in keywords:
        if word in description:
            recommended_foods.append((name,description))
            break
        
print(recommended_foods)
```

```python
# Print the recommended foods
print("Foods recommended for cooking with kids:")
for food in recommended_foods:
    name, _ = food  # Unpack the tuple
    print(f"- {name}")
```
