Show the number of recipes that require a “saucepan” to prepare that food. 

```python
# importing packages 
import pandas as pd
import json


# Loading the json files 
data = pd.read_json('recipes.json', lines = True)
data.head(5)
```

```python
# Function to check if "saucepan" is mentioned in the method steps
def requires_saucepan(method):
    return any("saucepan" in step.lower() for step in method)

# Apply the function to each row in the 'Method' column
saucepan_recipes = data[data['Method'].apply(requires_saucepan)]

# Count the number of recipes that require a "saucepan"
num_saucepan_recipes = len(saucepan_recipes)

print(f"Number of recipes that require a saucepan: {num_saucepan_recipes}")
```
