```python
import numpy as np 
import cv2 
```

```python
# Loading images 
img1 = cv2.imread('img1.jpg')
img2 = cv2.imread('img2.jpg')
```

```python
# convert images to greyscale 
gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
# Initialize the ORB detector algorithm
orb = cv2.ORB_create()
```

```python
# Detect keypoints and descriptors 
kp1, des1 = orb.detectAndCompute(gray1, None)
kp2, des2 = orb.detectAndCompute(gray2, None)
```

```python
# Initialize the Brute-Force Matcher
bf = cv2.BFMatcher()

# Match descriptors
matches = bf.match(des1, des2)
```

```python
# Sort matches by distance
matches = sorted(matches, key=lambda x: x.distance)
```

```python
# Draw best 25 matches 
matching_result = cv2.drawMatches(img1, kp1, img2, kp2, matches[:25], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
```

```python
# Save result
cv2.imwrite('orb-result.jpg', matching_result)
```

```python
# Display result (optional)
cv2.imshow('ORB Matches', matching_result)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```
